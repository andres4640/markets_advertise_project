# Analyzing Job interests

Now we will analyze in more detail how many participants are interested in each job role   

First we drop the rows without jobs interest and separate :

```python
#We delete a null values
jobs_clean = df.JobRoleInterest.dropna()
#Convert the column to string
jobs_inter = jobs_clean.astype(str)
#We lowercase the column and separete it by "," so it can be converted to a list.
jobs_inter = jobs_inter.str.lower().str.strip().str.split(",")
```

Second, we create a function that will clean the many job interest and will return a dictionary:

```python
job_dic = {}
def split_jobs(array):
    for row in array:
        #Takes out the start whitespace
        row = row.lstrip(' ')
        
        #Lowerscase the job roles they can be compared 
        row = row.lower()
        if row not in job_dic:       
            job_dic[row] = 1               
        else:
            job_dic[row] +=1

jobs_inter.apply(split_jobs)
print()
```

We convert the diccionary to a dataframe so we can clean it.

```python
job_df = pd.DataFrame.from_dict(list(job_dic.items()))
job_df = job_df.rename(columns={0:"Role", 1:"Count"})
```

```python
print(job_df)
print(job_df.shape)
```

```python
job_df = job_df[job_df.Count > 20]
job_df.reset_index(inplace=True, drop=True)
job_df
```

```python
job_df.sort_values(by="Count", ascending=False).plot.bar(y="Count", x="Role", figsize=(12,6))
```

During the data cleaning we lost some data, but this plot shows an approximate result that can represent the population.

# Spending money for learning analysis

We will analyze the 'MoneyForLearning' column that describes the amount of money the participants are willing to spent for courses. Our company sells subscriptions at a price of 59$ per month, this is why we want to know how much mone each student spends.

We are going to analyze only the top 4 countries with most highest absolute frequency in our sample: USA, India, UK and Canda. We do this because:

- There is a decent amount of data for each of this countries.
- English is an official language in this countries, there are more chances to target the right people and easier with our ads.


```python
df_new.MonthsProgramming.replace(0,1, inplace=True)
df_new["MoneySpend"] = df_new.MoneyForLearning / df_new.MonthsProgramming
```

Wen can see that there are a lot of participants that they want free courses (0$ willing to spend) and Null data (7736 rows). We will only keep the rows that are not nullls 

```python
df_new.MoneySpend.value_counts(dropna=False)
```

```python
temp_df  = df_new[df_new.MoneySpend.notnull()]
temp_df = temp_df[temp_df.CountryLive.notnull()]
```

We group the data by the country there are living and show the mean of the money they are willing to spend by month in learning to code.


```python
#Countries we want to analyze
countries_col = ["United States of America", "India", "United Kingdom", "Canada"]

#Group by country
countries_group = temp_df.groupby("CountryLive").agg({"MoneySpend":["mean","sum","max"]})["MoneySpend"] 
countries_group.loc[countries_col]
```

We can see that countries like UK and Canada with a higher GDP per capita than India should have a higher mean in money willing to spend by month. This could be caused by outliers or that we don't have enought data from this countries.

We can also see that there is a big difference between the mean and the maxium value, there could be some outliers.

# Dealing with outliers

Now we are going to make some box plots to visualize each country (USA, UK, India and Canada) distribution and watch for outliers.

```python
filter_countries = temp_df[temp_df.CountryLive.isin(countries_col)]

#Show outliers function
def show_outliers():
    
    plt.figure(figsize=(12,5))
    sns.boxplot(x="CountryLive", y="MoneySpend", data=filter_countries)
    plt.xticks(range(4),["USA", "UK", "India", "Canada"])
    plt.show()

    plt.figure(figsize=(12,5))
    sns.stripplot(x="CountryLive", y="MoneySpend", data=filter_countries, jitter=True)
    plt.show()
    
show_outliers()
```

We can see in the box plot that there are a lot of outlier in the data from USA participants. On the other hand we can see that the UK doesnt have much outliers. We should analyze this plots separated and be more meticulous.

We will elimninate the extreme outliers for the USA, this are the ones above 20000 in the column MoneySpend.

```python
filter_countries = filter_countries[filter_countries.MoneySpend <= 20000]

show_outliers()
```

```python
filter_countries.groupby("CountryLive").agg({"MoneySpend":["mean","max"]})["MoneySpend"] 
```

Now we can see outliers for each country. We can take the following actions:
- In USA, remove outliers that have more than 5500 in MoneySpend column.
- In India remove outliers with more than 2500.
- For Canada remove outliers with moren than 2500.

Before we take those actions, we should check the outliers and verify if thouse expenses are justified

```python
indian_out = filter_countries[(filter_countries.CountryLive == "India") & (filter_countries.MoneySpend >= 2500)]
indian_out
```

It is not specified in what the participants in India with more than 2500 dollars have spended the money. So we will drop out those columns because for lack of justification.

```python
filter_countries.drop(indian_out.index, inplace=True)
```

```python
show_outliers()
```

Now we will analyze the USA outliers:
- Participants with more than 4000 money spend

```python
usa_out = filter_countries[(filter_countries.CountryLive == "United States of America") & (filter_countries.MoneySpend >= 4000)]
usa_out[["AttendedBootcamp", "MoneyForLearning", "MonthsProgramming", "MoneySpend"]]
```

We are going to take out the those participants because some of them haven't attended to a BootCamp and other did attend, but they have attended only a few months to them and it is unrealistic that they had spend that much money in few months.

```python
filter_countries.drop(usa_out.index, inplace=True)
```

Now we will analyzze the Canada outliers, the outliers seems to be with a higher spending than 2500 dollars.

```python
canada_out = filter_countries[(filter_countries.CountryLive == "Canada") & (filter_countries.MoneySpend >= 2500)]
canada_out[["AttendedBootcamp", "MoneyForLearning", "MonthsProgramming", "MoneySpend"]]
```

We will remove this unrealistic value, they have attended only a few months to the bootcampo and it is unrealistic that they had spend that much money in two months. 

```python
filter_countries.drop(canada_out.index, inplace=True)
```

Now we should have a clean dataset:

```python
show_outliers()
```

```python
intervals = pd.interval_range(start=0, end=3500, freq=500)
print(pd.cut(filter_countries["MoneySpend"],bins=intervals).value_counts().sort_index())
```

# Choosing the two best markets

Now we are going to analyze the cleaned dataset and search for the two best markets to advertise in. We need to consider two factors:
- The company sells subscriptions at a price of 59 dollars per month.
- We need to search for markets with a considerable amout of posible customers

```python
filter_countries.groupby("CountryLive").agg({"MoneySpend":["mean","max"]})["MoneySpend"] 
```

```python
#First we filter by persons willing to pay 59 dollars or more for courses
markets = filter_countries[filter_countries.MoneySpend >= 59]
markets.groupby("CountryLive").agg({"MoneySpend":["mean","max", "count"]})["MoneySpend"] 
```

```python
print("=========Relative Frequency Table=================")
print(markets.CountryLive.value_counts(normalize=True)*100)
print("")
markets.CountryLive.value_counts().plot.bar(rot=45, title="Number of posible clients by country")
plt.show()
```

As we can see in the frequency table and in the plot, the best market to advertise is the USA. The second best country to advertise seems to be India.

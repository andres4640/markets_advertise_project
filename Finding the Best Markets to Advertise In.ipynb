# Analyzing Job interests

Now we will analyze in more detail how many participants are interested in each job role   

First we drop the rows without jobs interest and separate :

```python
#We delete a null values
jobs_clean = df.JobRoleInterest.dropna()
#Convert the column to string
jobs_inter = jobs_clean.astype(str)
#We lowercase the column and separete it by "," so it can be converted to a list.
jobs_inter = jobs_inter.str.lower().str.strip().str.split(",")

jobs_inter

```

Second, we create a function that will clean the many job interest and will return a dictionary:

```python
job_dic = {}
def split_jobs(array):
    for row in array:
        #Takes out the start whitespace
        row = row.lstrip(' ')
        
        #Lowerscase the job roles they can be compared 
        row = row.lower()
        if row not in job_dic:
            
            job_dic[row] = 1
                    
        else:
            job_dic[row] +=1

jobs_inter.apply(split_jobs)
print()
```

We convert the diccionary to a dataframe so we can clean it.

```python
job_df = pd.DataFrame.from_dict(list(job_dic.items()))
job_df = job_df.rename(columns={0:"Role", 1:"Count"})
```

```python
print(job_df)
print(job_df.shape)
```

```python
job_df = job_df[job_df.Count > 20]
job_df.reset_index(inplace=True, drop=True)
job_df
```

```python
job_df.sort_values(by="Count", ascending=False).plot.bar(y="Count", x="Role", figsize=(12,6))
```

During the data cleaning we lost some data, but this plot shows an approximate result that can represent the population.

 - Web development (full-stack, front-end and back-end) are the top 3 job roles that people are most interested in. There are approximate 10503 persons interested in web development.

# Densities Analysis

This analysis will be based in find in which countries are the majority of people living, each country is a possible market and we should find the two best countries to advertise in.

We will use the 'CountryLive' column for the analysis.

We will only consider the participants that that have responded the job role question 

```python
country_live = df[df["JobRoleInterest"].notnull()].copy()
country_live.reset_index(drop=True, inplace=True)
```

```python
rel_table = country_live.CountryLive.value_counts().sort_values(ascending=False)[:20]
absolut_table = country_live.CountryLive.value_counts(normalize=True).sort_values(ascending=False)[:20]*100

print("============Relative frequency table=============")
print(rel_table)
print("")
print("============Absolute frequency table==============")
print(absolut_table)
```

Based on this results the best two markets are the United States with 45.7% of the participants and India with 7.7%.
